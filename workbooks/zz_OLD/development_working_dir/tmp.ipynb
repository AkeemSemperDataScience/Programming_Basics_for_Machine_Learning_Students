```python
import pandas as pd
```

```python
df = pd.read_csv('assessments.csv')
df.head()
```

```python
df2 = df[df["Assessment Class 1"] == "RESIDENTIAL"]
df2 = df2[df2["Assessed Value"] > 50000]
df2 = df2[df2["Assessed Value"] < 2000000]
df2 = df2[df2["Suite"].isnull()]
df2.drop(columns=["Assessment Class 1", 
                  "Assessment Class 2", 
                  "Assessment Class 3", 
                  "Assessment Class % 1", 
                  "Assessment Class % 2", 
                  "Assessment Class % 3", 
                  "Ward", 
                  "Point Location", 
                  "Account Number", 
                  "Neighbourhood ID",
                  "Street Name", 
                  "House Number",
                  "Suite",
                  "Garage"], inplace=True)
df2.head()
```

```python
df2.info()
```

```python
df2.to_csv('residential2.csv', index=False)
```
