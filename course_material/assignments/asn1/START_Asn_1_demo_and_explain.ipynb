The test_asn1.py file can be used to do testing as you build. Each function will test something - the key part is the "assert" at the end. The assert statement tests is something is true, so you'll normally be testing something like:
<ul>
<li> Create a known answer (by hand, running code, etc.) </li>
<li> Run the function you want to test to get its answer. </li>
<li> Use assert to check if the known answer and the function answer match. </li>
</ul>

The idea here is to create a set of tests that cover the functionality of the program. As you work, you can run the tests to see if your code is working and if you've broken anything.

If we define the tests first, this is known as Test Driven Development (TDD). This is a good way to work, as it helps you focus on what you want to do and what you expect to get.
