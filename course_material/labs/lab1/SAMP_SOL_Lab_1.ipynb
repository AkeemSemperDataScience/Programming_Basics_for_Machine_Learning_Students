**1.  Write a statement that assigns total_coins with the sum of nickel_count and dime_count. The program snippet obtains nickel_count and dime_count from user input. (20%)**

For example, the output for 100 nickels and 200 dimes is 300. (20%)

wk2_lab1_1_testcase.GIF


```python
def coins():
    nickel_count = input("How many nickels?")
    dime_count = input("How many dimes?")
    total_coins = int(nickel_count) + int(dime_count)
    return total_coins

my_coins = coins()
print("You have", my_coins, "coins.")
```

**2. Write a statement that assigns cell_count with cell_count multiplied by 10. The program snippet obtians cell_count from user input.  (20%)**

If the input is 10, the output should be 100. (20%)

wk2_lab1_2_testcase.GIF

```python
cell_count = input("How many cells?")

def cellCount(input_num):
    return int(input_num) * 10

print("You have", cellCount(int(cell_count)), "cells.")
```

**3. Write a program that prompts the user for a measurement in meters and then converts it
to miles, feet, and inches. (20%)**

wk2_lab1_3_testcase.GIF

```python
input_num = input("What is the measure in meters?")

def converts(input_num):
    feet = float(input_num) * 3.28084
    miles = float(input_num) * 0.000621371
    inches = float(input_num) * 39.3701
    newline = "\n"
    text = f"{input_num} meters is {miles} miles. {newline}{input_num} meters is {feet} feet. {newline}{input_num} meters is {inches} inches."
    return (feet, miles, inches, text)

values = converts(input_num)
print(values[3])
```

**4. Write a program to compute how many gallons of paint are needed to cover the walls'
given square feet. Assume 1 gallon can cover 350.0 square feet. So gallons = the
square feet divided by 350.0. If the input is 250.0, the output should
be:0.714285714286. (20%)**

wk2_lab1_4_testcase.GIF

```python
input_num = input("How many square feet?")

def coverage(input_num):
    val = int(input_num) / 350
    txt = f"We need {val} gallons of paint."
    return (val, txt)

values = coverage(input_num)
print(values[1])
```

**5. Write a program that reads a five-digit positive integer and breaks it into a sequence of
individual digits. (20%)**

wk2_lab1_5_testcase.GIF

```python
input_num = input("Give me 5 digits!")

def splitter(input_num):
    # Error checking
    if len(input_num) != 5:
        print("Error - Please enter 5 digits!")
    if input_num.isnumeric() == False:
        print("Error - Please enter 5 Numerical digits!")
    # above ensures that the input is valid. 

    digits = []
    as_text = ""
    
    for char in input_num:
        digits.append(int(char))
        as_text += char + " "
    return (digits, as_text)

values = splitter(input_num)
print(values[1])
print(values[0])
```
