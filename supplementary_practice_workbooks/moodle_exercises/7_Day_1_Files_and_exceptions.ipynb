```python
# I want to swap values! How do I do it!

candy_types = ["Reese's", "Twix", "Snickers", "Coffee Crisp", "Big Turk"]

# a quick function to swap values of a certain index.
def swap(test_list, from_index, to_index):
    # saving the original value in a temp variable
    temp_value = test_list[from_index]
    # assign the orginal_value to the new_value
    test_list[from_index] = test_list[to_index]
    # assign the new_value to the temp_variable.
    test_list[to_index] = temp_value
    return test_list

candy_types = swap(candy_types, 0, 3)

print("Candy Power Rankings")
for index,candy in enumerate(candy_types):
    print(str(index + 1) + ": " + str(candy))
```

### Exception Handling

You can Raise exceptions which is one side of the coin and you can handle exceptions which is the other side of the coin.

```python
# Let's raise an error example.

def withdraw_funds(amount):
    if amount > 100:
        raise ValueError("You can't withdraw more than 100, dan. We know you.")
    return amount

# if it's less than 100 our program essentially works exactly as we expect it.
print(withdraw_funds(10))

# if it's greater we get an exception
funds = withdraw_funds(200)
```

```python
# We have a function here that defines withdrawing funds.

def withdraw_funds(amount):
    if amount > 100:
        raise ValueError("You can't withdraw more than 100, dan. We know you.")
    if amount < 0:
        raise ValueError("You can't withdraw negative funds unless you've reinvented math.")
    return amount

# this is how we handle it.
def person_withdrawing(amount):
    try:
        # try to see if this block works.
        funds = withdraw_funds(amount)
        print("You have withdrawn "+ str(funds))
    except ValueError as error_value:
        # if it fails then this block will execute.
        print("Sorry an error has occurred with your withdrawl: "+ str(error_value))

# three attempts
person_withdrawing(10) # this is successful

person_withdrawing(-100) # this is successful but tells user that there was error

person_withdrawing(10000) # this is successful but tells user that there was error but different than above.
```
