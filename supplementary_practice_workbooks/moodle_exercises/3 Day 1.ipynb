## Let's Take a look at some operators


```python
# checking if thing are equal.
first_num = 1
second_num = 2

is_equal = first_num == second_num
print("Is the first_num equal to the second_num")
print(is_equal)

if is_equal:
    print("Yeah Dan they are.")
else:
    print("No Dan they are not.")
```

```python
# let's check if it's greater than.
first_num = 1
second_num = 2

is_first_greater = first_num > second_num
print("is the first number greater than the second number")

print(is_first_greater)

# we can just use the operator right in the condition
if first_num > second_num:
    print("the first number is bigger")
```

```python
# Less than operator
first_num = 1
second_num = 2

is_first_less = first_num < second_num
print("Is the first number less than the second")
print(is_first_less)
```

```python
# Less than or equal to, let's see the difference.
first_num = 2
second_num = 2

print("is the first num less than the second")
print(first_num < second_num)

print("is the first num less than or equal to the second")
print(first_num <= second_num)

# just a note the greater or equal to is ">="
```

Note: you're never going to check if a string is greater than the other


```python
# not equal !=

print("Marshmallow" != "Gambit")
```

## Let's talk about elif and multiple alternatives!

```python
richter = float(input("What was the richter measurement of the earthquake? "))

if richter >= 8.0 :   # Handle the ‘special case’ first
   print("Most structures fall")
elif richter >= 7.0 :
   print("Many buildings destroyed")
elif richter >= 6.0 :
   print("Many buildings damaged, some collapse")
elif richter >= 4.5 :
   print("Damage to poorly constructed buildings")
else :    # so that the ‘general case’ can be handled last
   print("No destruction of buildings")

# !!! IMPORTANT !!!
# In an if, elif else example only one of the blocks is going to get executed
# not multiple.
# this is basically the main concept to understand.
```

## Let's take a look at logical operators

```python
# The and operator

liquid_temperature = int(input("How hot is your liquid? I will tell you the state... "))
FREEZING_POINT = 0
BOILING_POINT = 100

if liquid_temperature > FREEZING_POINT and liquid_temperature < BOILING_POINT:
    print("State is liquid")
elif liquid_temperature < FREEZING_POINT:
    print("State is solid")
elif liquid_temperature > BOILING_POINT:
    print("State is Vapor")
```

```python
# The Not operator

is_a_cat = True

if not is_a_cat:
    print("this is not a cat it's another animal")
else:
    print("this is a cat")

# The not operator basically just flips the truthiness of the variable or statement to the right of it.
```

```python
# The in operator in a string.

cat_names = "Ghost Gambit Marshmallow"

name = input("Enter your cat and see if it's dan's: ")

print("is it dan's cat?")
print(name in cat_names)
```
