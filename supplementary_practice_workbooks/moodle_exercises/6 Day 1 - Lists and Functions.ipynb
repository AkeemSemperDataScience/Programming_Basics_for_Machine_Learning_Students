### Functions Review

```python
# functions 

# what are we doing on the lines below we are declaring a function
def greet_person(first_name, last_name):
    return "Hello " + first_name + " " + last_name

# calling the function
print(greet_person("Dan", "Mouris"))

# using inputs
student_first_name = input("Please enter your first name: ")
student_last_name = input("Please enter your last name: ")

greeting = greet_person(student_first_name, student_last_name)

print(greeting)
```

```python
# making it all the way into it's own.
def greet():
    student_first_name = input("Please enter your first name: ")
    student_last_name = input("Please enter your last name: ")

    greeting = greet_person(student_first_name, student_last_name)

    print(greeting)

greet()
```

```python
# Difference between print and returning.

def sum_with_print(value_one, value_two):
    print(value_one + value_two)

print("Sum with print")
sum_with_print(2, 3)

#the above is going to print the value but you can't use the result.

def sum_with_return(value_one, value_two):
    return value_one + value_two

print("Sum with return")
result = sum_with_return(2,3)
print(result)
second_result = sum_with_return(result, 10)
print("the second sum is ")
print(second_result)
```

```python
# Scope example 1

def scope_example(name):
    print(name)
    # you can use "name" here because it's a parameter that has been defined.

print(scope_example("Ghost"))
# Below "name" is going to give an error as it only "lives" in the "scope_example"
print(name)
```

```python
first_name = "Dan"
# using globals. You want to avoid this. why? look below
def call_me_maybe():
    global first_name
    first_name = first_name + ", another one"
    print(first_name)

# this doesn't give you the same output!
call_me_maybe()
call_me_maybe()
call_me_maybe()

# try to avoid this!
```

### Lists What are they? Some Properties of lists

```python
# creating a list here's a few examples.
oiler_numbers = [97, 29, 18, 75]
cat_names = ["Ghost", "Gambit", "Marshmallow"]
oilers_did_they_win = [True, True, True]

oiler_stats = [
    ["Mcdavid", 48, 97],
    ["Draisaitl", 55, 100]
]

# we need to use square brackets not only to create them, but to access dem values.
print("Accessing values \n")
# How do I access the first value
print(oiler_numbers[0])
print(cat_names[1])

# Let's take a look at that last one.
print(oiler_stats[0]) # this returns the first list
# if we want to get the first value of inner list
print(oiler_stats[0][0])


# Let's do a replacement
print("Accessing values \n")
print("Before the change")
print(oilers_did_they_win)
# we
oilers_did_they_win[2] = False
print("After the change (and the ducks complete a miracle)")
print(oilers_did_they_win)


```

```python
# string immutability (immutability means it can't be reassigned or changed)

teacher_name = "Dan"

teacher_name[2] = "b"
print(teacher_name)
# this is going to give us an error because we can't assign it!

# this is a major difference to note with lists.
```

```python
# Let's talk about lists and Length
pokemon = ["Charizard", "Charmander", "Squirtle"]

# If you want to get the length
print(len(pokemon))

# if you try to access something that isn't in the list it's going to give you an error.
# below is going to give an index error
print(pokemon[6])
```

```python
# Let's talk about looping over that list.

pokemon = ["Charizard", "Charmander", "Squirtle", "Blastoise"]

# if we want to loop over all of these pokemon you can just use the for .. in ..
print("Without the index")
for creature in pokemon:
    print(creature)

# note you don't have the index above. Most use cases you won't need this.

# if you want to loop over and still have the index you just use enumerate.
print("\n With the index")
for index, creature in enumerate(pokemon):
    print("index "+str(index)+ " contains "+ creature)
```

```python
pokemon = ["Charizard", "Charmander", "Squirtle", "Blastoise"]

garys_pokemon = pokemon

# If you change one it's going to change it on both. This can be called a couple of things.
garys_pokemon[0] = "Snorlax"

print(pokemon)
print(garys_pokemon)

# If you really wanted to get a completely new list you need to do a deep copy.
import copy
chads_pokemon = copy.deepcopy(pokemon)
chads_pokemon[0] = "Meowth"
print(chads_pokemon)

# Note you're probably not going to need the last part but it's good to know.
```

### List Operations, what can you do with a list?

```python
# Appending Elements (adding elements on the end of a list.)
seasons = []
print(seasons)
# add a season to the seasons array.
seasons.append("Winter")
print(seasons)

# let's add one more
seasons.append("Summer")
print(seasons)


# Note you can see here that it's just going to print it on the end.

# Question can you append multiple in one line
seasons.append("Fall", "Spring")
# No this is going to give an error a TypeError
```

```python
# Let's take a look at insert
pokemon = ["Charizard", "Charmander", "Squirtle", "Blastoise"]

print(pokemon)
pokemon.insert(1, "Charmeleon")
# The second value is going to print out pokemon with the extra inserted pokemon.
print(pokemon)
```

```python
# Let's take a look at finding an item
pokemon = ["Charizard", "Charmander", "Squirtle", "Blastoise"]
print(pokemon)

# We want to find the index "Squirtle"
print("Where is squirtle")
found_index = pokemon.index("Squirtle")
print("he is at index "+str(found_index))
```

```python
# Let's talk about removing items
pokemon = ["Charizard", "Charmander", "Squirtle", "Blastoise"]
print(pokemon)
print(pokemon.pop()) # if you don't give an argument it just pops the last value
print(pokemon)

print(pokemon.pop(0)) # if we do give it a value it's going to pop it at that index.
print(pokemon)
```

```python
# Concatenation
# this is essentially just adding lists together

forwards = ["Mcdavid", "Draisaitl", "Hyman"]
defense = ["Nurse", "Barrie"]

# we are putting them altogether and showing the concatenated value
oilers = forwards + defense
print(oilers)
```

```python
# Wierd note You can replicate lists

# essentially it just repeats if you use the "multiplication" operator.
print([1,2,3]*5)

# Note it's very rare that you'll ever use this.
```

```python
# Equality and inequality

all_stars = ["Mcdavid", "Draisaitl"]
oilers_forwards = ["Mcdavid", "Draisaitl"]
ducks_forwards = ["Zegras", "Steele"]

forward_list=  ["Draisaitl", "Mcdavid"]

# below is equal. (using equality operator)
print(all_stars == oilers_forwards)

# below is not equal (using inequality operator)
print(oilers_forwards != ducks_forwards)


# Note they need to be in the same order
print(forward_list == oilers_forwards)
# this is false because they are not in the same order.
```

```python
# Sum Max and Min and sorting.
ages_of_players = [18, 39, 26, 27, 29]

# sum of the entire list
print(sum(ages_of_players))

# max of the list
print(max(ages_of_players))

# min of the list
print(min(ages_of_players))

# Let's sort the list
ages_of_players.sort()
print(ages_of_players)
```

```python
# Sum Max and Min and sorting.
ages_of_players = [18, 39, 26, 27, 29]
ages_of_players.sort()
print(ages_of_players)
# we know that it's sorted the lowest to highest.

# we want to get the youngest two players
youngest_two_players = ages_of_players[0:2]
print(youngest_two_players)

# let's get the oldest players
oldest_two_players = ages_of_players[-2:] # this is equivalent ages_of_players[3:5]
print(oldest_two_players) 


# some techniques
# up to second index
print(ages_of_players[:2])
# everything after and including the second index.
print(ages_of_players[2:])
```

```python
# Last example for this piece is using the list function

new_list = list("Dan")

# we can now change items in dan.
new_list[2] = "b"
print(new_list)

# here's how you turn it back into a string.
print("".join(new_list))
```

### Using Lists in functions

```python
# let's make our own sum function

def sum_with_list(our_values):
    total = 0
    for value in our_values:
        total = total + value

    return total

print(sum_with_list([1,2,3]))
print(sum_with_list([1,2,3,4,5]))
```

### Tables (lists in lists)

```python
oiler_stats = [
    ["Mcdavid", 48, 97],
    ["Draisaitl", 55, 100],
    ["Barrie", 10, 40]
]

for row in oiler_stats:
    row_string = ""
    for column in row:
        row_string = row_string +" " + str(column)

    print(row_string)

# if you have a list in a list you can use a for loop in a for loop to access all of the values.
```
